# Example script to work with Xarray

Working with netcdf files using xarray, some simple examples

```python
# import in sidebar (blue icon): netcdf4-python, regionmask (automatically imports xarray), cartopy


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.colors as colors
%matplotlib inline
import os, glob, re 
import math
import xarray as xr
import geopandas as gpd
import regionmask as regionmask
import netCDF4
import statsmodels.api as sm
```

```python
# Get the path of the data directory

datadir = os.path.join(os.getcwd(), 'sample-data')
datadir
```

```python
filepath = os.path.join(datadir, 'fwi_era5_daily_europe_2021.nc') 
var = 'fwinx' # select which variable you want to look at
# Open file and get key information
with xr.open_dataset(filepath) as ds:
    print(str(list(ds.keys()))) # the names of all the variables
    print(str(list(ds.dims))) # time, lat, lon are the dimensions
    fwi_2020 = ds[var] # extract one variable from the data
```

```python
# look at the fire weather index data key information 

fwi_2020
```

```python
# calculate the mean through time and make a simple plot with xarray

fwi_2020_mean = fwi_2020.mean(dim='time')
fwi_2020_mean.plot()

# you can make more complex plots using matplotlib and cartopy (e.g. you can plot country boundaries with cartopy)
```

```python
# get maximum at each grid cell and plot


fwi_2020_max = fwi_2020.max(dim='time')
fwi_2020_max.plot()
```

```python
# take average over land and plot year by year evolution 

# set up plot
fig, ax = plt.subplots()

# plot daily timeseries 
fwi_2020_timeseries_daily = fwi_2020.mean(dim=['latitude', 'longitude'])
fwi_2020_timeseries_daily.plot(ax=ax, label='daily timeseries')

# take the mean over each month and plot 
fwi_2020_timeseries_monthly = fwi_2020_timeseries_daily.resample(time='M', label='right').mean()
fwi_2020_timeseries_monthly.plot(ax=ax, label='monthly timeseries')

plt.legend()

# FWI peaks in the summer 
```

```python
# calculate trend in time in Brussels over the first 6 months

def fit_trend(data):
    x=range(len((data.time)))
    y=data.values
    fit=sm.OLS(y, sm.add_constant(x)).fit()    
    return fit, x, y

# select grid cell of brussels and take only the first six months
fwi_2020_Brussels = fwi_2020.sel(latitude=50.85045, longitude=4.34878, method='nearest').sel(time=slice('2021-01-01', '2021-07-01')) 

fit, x, y = fit_trend(fwi_2020_Brussels)
fit.summary()
```

```python
fit.params # first is intercept, second is slope
```

```python
plt.scatter(x,y)
plt.axline([0, fit.params[0]], slope=fit.params[1], ls='--')

# plot
```
